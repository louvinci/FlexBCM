### Vanilla\FFTConv\OctConv计算量比较

#### 小卷积核

```python
import torch
import numpy as np

def getOctLayer(layer,alpha=0.25):
    N,M,R,C,K,S,P = layer[0],layer[1],layer[2],layer[3],layer[4],layer[5],layer[6]
    alpha_in,alpha_out  = alpha,alpha
    octlayer=[]
    #h2h
    octlayer.append([ N*(1-alpha_in), M*(1-alpha_out), R,C,K,S,P])
    #h2l
    octlayer.append([ N*(1-alpha_in), M*alpha_out,     R/2,C/2,K,S,P])
    #l2h
    octlayer.append([ N*alpha_in,     M*(1-alpha_out), R/2,C/2,K,S,P])
    #l2l
    octlayer.append([ N*alpha_in,     M*alpha_out,     R/2,C/2,K,S,P])
    return octlayer

def next_power2(x):
    # return math.pow(2, math.ceil(math.log2(x)))
    return int(2 ** np.ceil(np.log2(x)))

# Conv layer ops 1-dim list size:7
def Conv_OPs(layer):
    n,m,h,w,k,s,p = layer[0],layer[1],layer[2],layer[3],layer[4],layer[5],layer[6]
    r = (h+2*p-k)/s+1
    c = (w+2*p-k)/s+1
    mem = h*w*n+ r*c*m + k*k*n*m
    #r,c = h,w
    return 2*(n*m*r*c*k*k),mem

# Oct layer ops, layer:2-dim list 4x7
def Oct_layerOPs(layers):
    assert len(layers) == 4,'should contain four sub-layers'
    nop,mem =0,0
    for j in range(4):
        t_nop, t_mem = Conv_OPs(layers[j])
        nop += t_nop
        mem += t_mem
    return nop,mem

# n:input channels, m:output channels,h,w height = width, k:kernel size, s:stride, p:padding
def FFT_layerOPs(layer,pre_w=True):
    n,m,h,w,k,s,p = layer[0],layer[1],layer[2],layer[3],layer[4],layer[5],layer[6]
    # if k == 1:
    #     k = 3
    # pad_k = next_power2(k)
    mem = 0
    pad_k = 8 if k <=3 else 16 # the FFT size可以参见FPGA18来确定如何选取
    # tile size = pad_k,when using OaA
    tile_size = pad_k -k+1 #(LxL)
    tiles = np.ceil((h/s)/tile_size)
    fft_ops, harda_ops = 0,0
    tile_comlex =  (pad_k * pad_k -4) /2 # 复数部分，除2因为共轭对称
    tile_real = 4 # 四个实数部分
    harda_ops = 2*(tile_comlex *3 + tile_real)*n*m * tiles * tiles # 复数乘法优化为3个乘法,tiles也需要相加因此乘2
    outtile_fft = intile_fft = (2*pad_k*pad_k * np.log2(pad_k)) *3
    weight_fft = 0 if pre_w else 2*pad_k*pad_k * np.log2(pad_k) *3

    fft_ops = (intile_fft * n + outtile_fft * m ) * tiles*tiles+ weight_fft*n*m
    
    mem = pad_k*pad_k * tiles*tiles*(n+m ) + pad_k*pad_k*n*m
    return (fft_ops+ harda_ops),mem


```

```python
layer1 = [ 3,   64, 224, 224,   3,   1,   1]
layer2 = [ 64,  64, 224, 224,   3,   1,   1]
#layer = [ 16,  48, 56, 56,   1,   1,   1]
fft_ops = FFT_layerOPs(layer1)[0]  + FFT_layerOPs(layer2)[0]
spatial_ops = Conv_OPs(layer1)[0] + Conv_OPs(layer2)[0]
print(fft_ops,' ',spatial_ops,' ',spatial_ops /fft_ops )
```

```python
skernel = [1,3]
width = [224, 112, 112,  56,  56,  28,  28,  14,14]
scin =  [  3,  32,  32,  64, 128, 128, 256, 256,512]
scout = [ 32,  32,  64, 128, 128, 256, 256, 512,512]
alpha = 0.375
```

```python
for i in range(len(skernel)):
    k = skernel[i]
    pad = 1 if k <= 3 else 2 if k==5 else 3
    stride = 1 if k <= 3 else 2
    print('========================k={}========================'.format(k))
    for j in range(len(scin)):
        cin,cout = scin[j],scout[j]
        h = w = width[j]
        layer = [cin,cout,h,w,k,stride,pad]
        spatial_ops = Conv_OPs(layer)[0]
        fft_ops = FFT_layerOPs(layer)[0]
        oct_ops = Oct_layerOPs(getOctLayer(layer,alpha))[0]
        print('<cin,cout,h,w,k,s,p>:{0} {1} {2} {3} {4} {5} {6}'.format(cin,cout,h,w,k,stride,pad))
        print(spatial_ops,fft_ops,oct_ops)
        print('ratio: {0},{1:.2f},{2:.2f}'.format(1,spatial_ops/fft_ops,spatial_ops/oct_ops))
    
```

```python
flag1 = [0]*13
flag2 = [0]*4+[1]*9

print(len(vgg16D),flag1,flag2)
```

ResNet

```python
import numpy as np
def ctcfunc(layers,flag,bytes=4):
    assert len(layers) == len(flag),'size should be same'
    ctc = []
    l_mem,l_mac = 0,0
    for i in range(len(layers)):
        if flag[i] ==0:
            l_mac,l_mem = Conv_OPs(layers[i])
            #print(l_mac,l_mem)
        else: 
            l_mac,l_mem = FFT_layerOPs(layers[i])
        l_ctc = (l_mac / (l_mem*1000*bytes)) # GOP/MB
        ctc.append(l_ctc)
    return ctc
```

```python
re1 = ctcfunc(vgg16D,flag1)
re2 = ctcfunc(vgg16D,flag2)
print(re1,'\n',re2)
print(np.var(re1),np.var(re2))
```
